```python
import matplotlib.pyplot as plt
import pandas as pd
```

Cultivated  in Lakh Ha

```python
df = pd.read_csv('millet_area.csv')

years = df['Year'].unique().tolist()
Ragi_area = df['Ragi'].tolist()
Jowar_area = df['Jowar'].tolist()
Bajra_area = df['Bajra'].tolist()
```

Production in MT

```python
df2 = pd.read_csv('millet_prod.csv')

Bajra_prod = df2['Bajra'].tolist()
Jowar_prod = df2['Jowar'].tolist()
Ragi_prod = df2['Ragi'].tolist()
```

Export in MT

```python
df3 = pd.read_csv('millet_export_qt.csv')

Bajra_export = df3['Bajra'].tolist()
Jowar_export = df3['Jowar'].tolist()
Ragi_export = df3['Ragi'].tolist()
```

Import in MT

```python
df4 = pd.read_csv('millet_import_qt.csv')

Ragi_import = df4['Ragi'].tolist()
Jowar_import = df4['Jowar'].tolist()
Bajra_import = df4['Bajra'].tolist()
```

Nutritive values of Rice, Wheat and Millets

```python
n1 = pd.read_csv('rice_cal.csv')
n2 = pd.read_csv('wheat_cal.csv')
n3 = pd.read_csv('millets_cal.csv')

year = n1['Year'].tolist()
rice_calories = n1['Value'].tolist()
wheat_calories = n2['Value'].tolist()
millets_calories = n3['Value'].tolist()
```

```python
plt.plot(year, rice_calories, label='Rice Calories', color='blue')
plt.plot(year, wheat_calories, label='Wheat Calories', color='red')
plt.plot(year, millets_calories, label='Millet Calories', color='green')

plt.title('Daily Nutritive consumption of Wheat, Rice and Millet')
plt.xlabel('Year')
plt.ylabel('Food supply (kcal/capita/day)')
plt.legend(['Rice', 'Wheat', 'Millet'])
```

Area Plot

```python
plt.plot(years, Ragi_area, color='red', label='Ragi')
plt.plot(years, Bajra_area, color='blue', label='Bajra')
plt.plot(years, Jowar_area, color='green', label='Jowar')

plt.xlabel('Years')
plt.xticks(rotation=90)
plt.ylabel('Area (in Lakh-Ha)')
plt.title('Yearly Area under cultivation of millets')

plt.ylim(top=100)

plt.legend(['Ragi', 'Bajra', 'Jowar'], loc='upper right')

plt.show()
```

Production Plot

```python
plt.plot(years, Ragi_prod, color='red', label='Ragi Production')
plt.plot(years, Ragi_import, color='blue', label='Ragi Import')
plt.plot(years, Ragi_export, color='green', label='Ragi Export')

plt.xlabel('Years')
plt.xticks(rotation=90)
plt.ylabel('Quantity (in Ton)')
plt.title('Ragi')

plt.ylim(top=120000)

plt.legend(['Ragi Production', 'Ragi Import', 'Ragi Export'], loc='upper right')

plt.show()
```

Export plot

```python
plt.plot(years, Jowar_prod, color='red', label='Jowar production')
plt.plot(years, Jowar_import, color='blue', label='Jowar Import')
plt.plot(years, Jowar_export, color='green', label='Jowar Export')

plt.xlabel('Years')
plt.xticks(rotation=90)
plt.ylabel('Quantity (in Ton)')
plt.title('Jowar')

plt.ylim(top=60000)

plt.legend(['Jowar production', 'Jowar Import', 'Jowar Export'], loc='upper right')

plt.show()
```

Import Plot

```python
plt.plot(years, Bajra_prod, color='red', label='Bajra Production')
plt.plot(years, Bajra_import, color='blue', label='Bajra Import')
plt.plot(years, Bajra_export, color='green', label='Bajra Export')

plt.xlabel('Years')
plt.xticks(rotation=90)
plt.ylabel('Quantity (in Ton)')
plt.title('Bajra')

plt.ylim(top=300000)

plt.legend(['Bajra Production', 'Bajra Import', 'Bajra Export'], loc='upper right')

plt.show()
```
